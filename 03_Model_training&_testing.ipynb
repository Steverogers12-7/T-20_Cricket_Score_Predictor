```python
import pandas as pd
df=pd.read_csv('./Dataset_04.csv')
```

```python
data=df
data
```

```python
x=data['balls_left']
```

```python
phase=[]
for i in x:
    if i>90:
        phase.append('powerplay')
    elif i<30:
        phase.append('death')
    else: phase.append('middle')
phase
```

```python
agression=[]
x=data['crr']
y=data['wickets_left']
```

```python
data['aggression'] = data['crr'] * data['wickets_left']
```

```python
data['phase']=phase
```

```python
data.info()
```

```python
data['progress'] = (120 - data['balls_left']) / 120
data['runs_possible'] = data['crr'] * (data['balls_left'] / 6)
data.info()
```

```python
data.to_csv('Dataset_04.csv',index=False)
```

```python
X=data.drop(columns=['runs_x']) #Input Features
Y=data['runs_x'] #Output Features
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city','phase'])],
                      remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2', XGBRegressor(
        n_estimators=800,
        learning_rate=0.05,
        max_depth=8,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=1
    ))
])
```

```python
#Training data over algorithm.
pipe.fit(X_train,Y_train)
```

```python
#prediction over testing data.
Y_pred=pipe.predict(X_test)
```

```python
print(r2_score(Y_test,Y_pred))
print(mean_absolute_error(Y_test,Y_pred))
```

```python
from sklearn.model_selection import cross_val_score

# 5-fold cross validation
scores = cross_val_score(pipe, X_train, Y_train, cv=5, scoring='r2')

print("Cross-validation R2 scores:", scores)
print("Mean CV R2:", scores.mean())
print("Std deviation:", scores.std())
```

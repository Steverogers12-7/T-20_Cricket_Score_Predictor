```python
import pandas as pd
df=pd.read_csv('./Dataset_04.csv')
```

```python
df
```

```python
# Now Doing train-test split. 
#This is Regression problem.
#output feature= runs_x(how much scored at the end).
```

```python
df.sample(2)
```

```python
X=df.drop(columns=['runs_x'])
Y=df['runs_x']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])],
                      remainder='passthrough')
```

```python
# we make pipeline that handles all columns.
pipe=Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
#Training  training data over algorithm 
pipe.fit(X_train,Y_train)
```

```python
#prediction over testing data
Y_pred=pipe.predict(X_test)
```

```python
#Regression metrics value
```

```python
print(r2_score(Y_test,Y_pred))
print(mean_absolute_error(Y_test,Y_pred))
```
